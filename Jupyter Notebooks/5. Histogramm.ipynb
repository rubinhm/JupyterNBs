# 5. Histogramme

Histogramme dienen allgemein der Veranschaulichung von Häufigkeitsverteilungen oder Wahrscheinlichkeiten. Im letzten Notebook (Nr. 4) hatten wir ein Histogtamm verwendet, um die Häufigkeit der Messwerte eines Zwei-Qubit Registers darzustellen. Im folgenden stellen wir diese Sequenz noch einmal zusammenhängend dar:

```python
from qiskit import QuantumCircuit, execute, BasicAer
from qiskit.visualization import plot_histogram
```

```python
qc=QuantumCircuit(2, 2)
qc.h([0,1])
qc.measure([0, 1], [0, 1])
backend = BasicAer.get_backend('qasm_simulator')
job=execute(qc, backend, shots=10000)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```

Schauen wir uns die Variable *counts* mithilfe der *print*-Funktion genauer an:

```python
print (counts)
```

`counts` ist eine Python-Liste (eine Art mehrkomponentiger Zeichenkette). Diese Liste können wir auch von Hand erstellen, z. B. also {'00': 123, '01': 87, '10': 5, '11': 23} und diese als Argument plot_histogramm() übergeben:

```python
plot_histogram({'00': 123, '01': 87, '10': 5, '11': 23})
```

Wir könnten aber auch z. B. Wahlergebnisse darstellen:

```python
plot_histogram({'AfD':22, 'CDU': 22, 'Linke': 9, 'Grüne': 13, 'SPD':16}, sort='desc', title='Ergebnisse der Bundestagswahl 2025 in %')
```

Beachten Sie die Reihenfolge, sie kann mit dem Argument *sort* gesteuert werden. Siehe die [Referenz](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.plot_histogram) dazu.

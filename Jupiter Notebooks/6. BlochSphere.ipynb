# 6. Die Blochkugel

Die Blochkugel ist uns schon im 1. Notebook begegnet, als es um die Darstellung von Qubit-Zuständen ging. Mit der Blochkugel kann der Zustand **eines** Qubits anschaulich dargestellt durch einen Zustandsvektor im Raum mit dem Betragsquadrat $1$ dargestellt werden. Die Enden dieses Vektors liegen deshalb auf einer Kugel mit Radius $r=1$, der Blochkugel. Die folgende Abbildung zeigt dies noch einmal für den Fall, dass in $z$-Richtung gemessen wird (Messbasis $z$):

Dieser Zustandsvektor könnte als dreikomponentiger Vektor dargestellt werden. Und zwar entweder in kartesischen Koordinaten ($x, y, z$), wie im folgenden Beispiel:

```python
from qiskit.visualization import plot_bloch_vector
from math import sqrt, pi
```

```python
plot_bloch_vector([0, 0, 1],title='Bloch-Vektor in kartesischen Koordinaten')
```

oder in Polarkoordinaten ($r, \theta, \varphi$):

```python
plot_bloch_vector([1, pi/2, pi/2], coord_type='spherical', title='Bloch-Vektor in Polarkoordinaten')
```

Der Zustandsvektor wird auch dann ohne Fehlermeldung gezeichnet, wenn der Vektor nicht auf $1$ normiert ist. Dies entspricht allerdings keinem reinen quantenmechanischen Zustand. Der *statevector_simulator* (s. Notebook 3) würde in diesem Fall eine Fehlermeldung ausgeben. Probieren Sie verschiedene Längen aus!

In beiden Fällen sind die drei Komponenten **reelle Zahlen**. Hier stellt sich die Frage, weshalb die Amplituden in der Quantenmechanik i. A. **komplexe Zahlen** sein sollen?





Nun das Merkwürdige ist, dass der Zustandsvektor eines Qubits als dreikomponentiger Vektor dargestellt werden kann und und ein Qubit-Zustand einem beliebigen Punkt auf der Blochkugel entsprechen kann. Bei einer Messung (hier in $z$-Richtung) hingegen sind nur zwei Ergebnisse möglich, nämlich $0$ oder $1$. Hinzu kommt, dass man diese Ergebnisse nicht mit Sicherheit vorhersehen kann. Sie sind rein zufällig (quantenmechanischer Zufall bzw. reiner oder objektiver Zufall). Die Messwahrscheinlichkeit lässt sich hingegen berechnen. Sie hängt von den Anteilen des Zustandsvektors in $|0\rangle$-Richtung und in $|1\rangle$-Richtung ab.

In unserem ersten Beispiel mit den kartesischen Koordinaten, befindet sich der Bloch-Vektor (Zustandsvektor) bereits im Zustand $|0\rangle$. Eine Messung liefert dann mit Sicherheit das Messergebnis $0$. Im zweiten Beispiel mit den Polarkoordinaten, liegt der Bloch-Vektor auf der $y$-Achse. Er hat keinen Anteil in $|0\rangle$-Richtung oder in $|1\rangle$-Richtung. Die beiden Messergebnisse $0$ und $1$ sind deshalb gleich wahrscheinlich, nämlich $50 \%$ bzw. $0.5$. 

In der Zustandsfunktionsollen diese Messwahrscheinlichkeiten enthalten sein, weil das das einzige ist, was wir über diesen Zustand erfahren können. Deshalb könnte man den Zustand folgendermassen schreiben:

$$|q_0\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle$$

Dies, weil nach der Born'schen Regel die Quadrate der Amplituden den Messwahrscheinlichkeiten entsprechen sollen.

Nun kann der Blochvektor aber irgendwo in der $xy$-Ebene liegen, die Messwahrscheinlichkeiten wären dieselben. Dennoch unterscheidet sich der Zustandsvektor von oben in $y$-Richtung von einem in $x$-Richtung. Wie lassen sich also diese Zustände mathematisch unterscheiden?

Eine Möglichkeit besteht darin, komplexe Zahlen einzuführen. Dann kann die $xy$-Ebene als komplexe Zahlenebene aufgefasst werden mit z. B. $x$ als reelle Achse und $y$ als imaginäre Achse. Damit lässt sich der oben abgebildete Zustand eindeutig schreiben als:
$$|q_0\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{i}{\sqrt{2}}|1\rangle$$

Liegt der Blochvektor irgendwo in der $xy$-Ebene, müssen wir seine Lage durch den Winkel $\varphi$ beschreiben. Das ginge dann folgendermassen:

$$
|q\rangle = \frac{1}{\sqrt{2}}|0\rangle + e^{i\phi}\frac{1}{\sqrt{2}}|1\rangle
$$

Für eine beliebige Lage muss schliesslich noch der Winkel $\theta$ berücksichtigt werden. Wie das gemacht werden muss, wurde in Notebook 1 dargelegt. Wir erhalten schliesslich die allgemeine Form für einen Bloch-Vektor:

$$
|q\rangle = \cos{\tfrac{\theta}{2}}|0\rangle + e^{i\phi}\sin{\tfrac{\theta}{2}}|1\rangle
$$

Bei der Berechnung der Messwahrscheinlichkeit verschwindet der Term $e^{i\phi}$, weil $|e^{i\phi}|^2=e^{i\phi}\cdot e^{-i\phi}=1$ ist.

Der Phasenwinkel $\varphi$ hat somit keinen Einfluss auf die Messwahrscheinlichkeit. Man nennt ihn oft auch die *globale Phase*. Dennoch darf diese *globale Phase* bei der Betrachtung von Quantenregistern nicht ausser Acht gelassen werden. Wie bei der Überlagerung Wellen ist sie entscheidend für das Verhalten  des Registers, also z. B. ob die Qubits konstruktiv oder destruktiv interferieren.
